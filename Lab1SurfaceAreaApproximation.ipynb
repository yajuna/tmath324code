This code explains the surface area compuation from Section 14.5. We already know the surface area of $z=f(x,y)$ over the region $R$ is $S=\iint_R\sqrt{f^2_x+f^2_y+1}\ dA$. This code plots vectors used in the derivation of the formula.

```python
## import packages, and conventional abbrev.
import numpy as np
import scipy as sp
import matplotlib 
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d # for 3D coordinates
```

Plot a 3D surface, $z= -x^2-y^2 + 4$

```python
x = np.outer(np.linspace(-2, 2, 30), np.ones(30)) # outer product 
y = x.copy().T # transpose

# if change surface, change this
z = -x ** 2 - y ** 2 + 4 # define surface 

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface $z = -x^2 -y^2 + 4 $')
plt.show()
```

Add vectors $\mathbf{u}=\Delta x\mathbf{i}+ 0\mathbf{j} + f_x \Delta x\mathbf{k}$ and $\mathbf{v} = 0\mathbf{i} + \Delta y\mathbf{j} + f_y\Delta y\mathbf{k}$. These vectors are parallel to the $xz$-plane, and $yz$-plane respectively. 

```python
# take Delta x = 0.5, Delta y = 0.3, vect are u and v, at (1,-1,2)
# vectProj are projections of u and v on the xy-plane.

import numpy as np
######  TYPO FIXED
vect = np.array([[1,-1,2,1.5,-1,1],[1,-1,2,1,-0.7,2.6]])
# vect = (initial point, end point) = (intial point, initial point + u)

vectProj = np.array([[1,-1,0,1.5,-1,0],[1,-1,0,1,-0.7,0]])

X, Y, Z, U, V, W = zip(*vect)
A, B, C, D, E, F = zip(*vectProj)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.quiver(A, B, C, D, E, F)
ax.plot_surface(x, y, z,cmap='viridis', edgecolor='none')
ax.set_title('Surface $z = -x^2 -y^2 + 4$, with vectors $u$ and $v$ at (1,-1,2)')
plt.show()
```
